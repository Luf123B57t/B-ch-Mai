```python
table_map = {
    "chartevents": chartevent,
    "discharge": discharge,
    "labevents": labevent,
    "microbiology": microbiologyevent,
    "procedureevents": procedureevent,
    "radiology": radiologyevent,
}

keyword_map = {
    "Ớn lạnh": ["chills", "rigors", "shivering", "feeling cold", "cold"],
    "Tiểu gấp": ["urinary urgency", "urgency"],
    "Tiểu nhiều lần": ["urinary frequency", "frequent urination"],
    "Tiểu buốt": ["dysuria", "painful urination", "burning on urination"],
    "Đau hông sườn": ["flank pain", "cva tenderness"],
    "Đau/ấn đau vùng trên xương mu": ["suprapubic pain", "suprapubic tenderness"],
}

central_line_locations = [
    "Left IJ", "Right IJ",
    "Left Subclavian", "Right Subclavian",
    "Left Femoral", "Left Femoral.",
    "Right Femoral", "Right Femoral."
]

def get_variable_data(variable_name, subject_id=None, hadm_id=None, stay_id=None):
    mapping = variable_mapping[
        variable_mapping["variable_name"] == variable_name
    ].copy()

    results = []

    for _, row in mapping.iterrows():
        df = table_map[row["source_table"]].copy()

        if subject_id is not None and row["subject_id_column"] in df.columns:
            df = df[df[row["subject_id_column"]] == subject_id]

        if hadm_id is not None and row["hadm_id_column"] is not None and row["hadm_id_column"] in df.columns:
            df = df[df[row["hadm_id_column"]] == hadm_id]

        if stay_id is not None and row["stay_id_column"] is not None and row["stay_id_column"] in df.columns:
            df = df[df[row["stay_id_column"]] == stay_id]

        if row["lookup_operator"] == "==" and row["lookup_column"] is not None:
            df = df[df[row["lookup_column"]] == row["lookup_value"]]

        elif row["lookup_operator"] == "in_list":
            if row["lookup_column"] == "location" and row["lookup_value"] == "__CENTRAL_LINE__":
                df = df[df["location"].isin(central_line_locations)]

        elif row["lookup_operator"] == "keyword_search":
            keywords = keyword_map.get(variable_name, [])
            if "text" in df.columns and keywords:
                pattern = "|".join(keywords)
                df = df[df["text"].str.contains(pattern, case=False, na=False, regex=True)]

        if row["secondary_lookup_column"] is not None:
            col2 = row["secondary_lookup_column"]
            val2 = row["secondary_lookup_value"]
            if col2 in df.columns:
                df = df[df[col2] == val2]

        df = df.copy()
        df["mapped_variable_name"] = row["variable_name"]
        df["mapped_source_table"] = row["source_table"]
        df["mapped_value_column"] = row["value_column"]
        df["mapped_time_column"] = row["time_column"]
        df["mapped_endtime_column"] = row["endtime_column"]

        results.append(df)

    if not results:
        return pd.DataFrame()

    result = pd.concat(results, ignore_index=True, sort=False)
    result = result.drop_duplicates().reset_index(drop=True)
    return result
```

```python
import pandas as pd
import numpy as np

# =========================
# 1) Helper
# =========================
def _to_datetime_safe(s):
    return pd.to_datetime(s, errors="coerce")

def _contains_any(series, keywords):
    pattern = "|".join(keywords)
    return series.astype(str).str.contains(pattern, case=False, na=False, regex=True)

# =========================
# 2) VAP check for one subject
# =========================
def check_vap_subject(subject_id, verbose=True):
    # ---------- A. Chartevents ----------
    ce = chartevent[chartevent["subject_id"] == subject_id].copy()
    if not ce.empty:
        ce["charttime"] = _to_datetime_safe(ce["charttime"])

    # Nhiệt độ > 38C
    temp_c = ce[(ce["itemid"] == 223762) & (ce["valuenum"] > 38)].copy()

    # Ho: Cough Effort weak/strong
    cough = ce[
        (ce["itemid"] == 223991) &
        (ce["value"].astype(str).str.lower().isin(["weak", "strong"]))
    ].copy()

    # Đờm mủ: Sputum Color != Clear
    sputum = ce[
        (ce["itemid"] == 224370) &
        (~ce["value"].astype(str).str.lower().eq("clear"))
    ].copy()

    # Nhịp thở > 20
    rr = ce[(ce["itemid"] == 230042) & (ce["valuenum"] > 20)].copy()

    # SpO2 < 94
    spo2 = ce[(ce["itemid"] == 220277) & (ce["valuenum"] < 94)].copy()

    # Có hỗ trợ oxy: O2 Delivery Device != none
    oxygen_support = ce[
        (ce["itemid"] == 226732) &
        (~ce["value"].astype(str).str.lower().eq("none"))
    ].copy()

    # Rale / ran / âm phổi bất thường
    lung_sound_keywords = [
        "bronchial", "crackles", "diminished",
        "exp wheeze", "ins/exp wheeze", "insp wheeze",
        "rhonchi", "tubular"
    ]
    lung_sounds = ce[
        ce["itemid"].isin([223986, 223987, 223988, 223989]) &
        _contains_any(ce["value"], lung_sound_keywords)
    ].copy()

    # ---------- B. Radiology ----------
    rad = radiologyevent[radiologyevent["subject_id"] == subject_id].copy()
    pneumonia_keywords = [
        "pneumonia", "pna", "bronchopneumonia", "consolidation",
        "infiltrate", "infiltration", "opacity", "opacities",
        "ground glass", "ggo", "air bronchogram", "patchy",
        "cloudy", "hazy"
    ]
    if not rad.empty:
        imaging = rad[
            (rad["field_name"].astype(str).str.upper() == "FINDINGS") &
            _contains_any(rad["field_value"], pneumonia_keywords)
        ].copy()
    else:
        imaging = pd.DataFrame()

    # ---------- C. Microbiology ----------
    micro = microbiologyevent[microbiologyevent["subject_id"] == subject_id].copy()
    if not micro.empty:
        micro["charttime"] = _to_datetime_safe(micro["charttime"])

    resp_specimens = [
        "BRONCHIAL WASHINGS",
        "BRONCHOALVEOLAR LAVAGE",
        "PLEURAL FLUID",
        "SPUTUM"
    ]

    resp_culture = micro[
        micro["spec_type_desc"].astype(str).isin(resp_specimens) &
        micro["org_name"].notna()
    ].copy()

    blood_culture = micro[
        micro["spec_type_desc"].astype(str).eq("BLOOD CULTURE") &
        micro["org_name"].notna()
    ].copy()

    # ---------- D. Procedureevents: thở máy ----------
    proc = procedureevent[procedureevent["subject_id"] == subject_id].copy()
    if not proc.empty:
        proc["starttime"] = _to_datetime_safe(proc["starttime"])
        proc["endtime"] = _to_datetime_safe(proc["endtime"])

    vent = proc[proc["itemid"] == 225792].copy()

    # =========================
    # 3) Check by stay_id
    # =========================
    stay_ids = sorted(ce["stay_id"].dropna().unique().tolist()) if not ce.empty else []

    results = []

    for stay_id in stay_ids:
        ce_stay = ce[ce["stay_id"] == stay_id].copy()
        hadm_ids = ce_stay["hadm_id"].dropna().unique().tolist()
        hadm_id = hadm_ids[0] if len(hadm_ids) > 0 else np.nan

        temp_stay = temp_c[temp_c["stay_id"] == stay_id]
        cough_stay = cough[cough["stay_id"] == stay_id]
        sputum_stay = sputum[sputum["stay_id"] == stay_id]
        rr_stay = rr[rr["stay_id"] == stay_id]
        spo2_stay = spo2[spo2["stay_id"] == stay_id]
        oxygen_stay = oxygen_support[oxygen_support["stay_id"] == stay_id]
        lung_stay = lung_sounds[lung_sounds["stay_id"] == stay_id]

        # triệu chứng lâm sàng
        symptom_hits = {
            "fever": not temp_stay.empty,
            "cough": not cough_stay.empty,
            "purulent_sputum": not sputum_stay.empty,
            "tachypnea": not rr_stay.empty,
            "low_spo2": not spo2_stay.empty,
            "oxygen_support": not oxygen_stay.empty,
            "abnormal_lung_sounds": not lung_stay.empty,
        }
        symptom_count = sum(symptom_hits.values())

        # imaging theo hadm_id nếu có thể
        # radiology_detail bạn đưa không có hadm_id/time, nên check ở mức subject
        imaging_positive = not imaging.empty

        # microbiology theo hadm_id nếu có
        if not np.isnan(hadm_id):
            resp_culture_stay = resp_culture[resp_culture["hadm_id"] == hadm_id]
            blood_culture_stay = blood_culture[blood_culture["hadm_id"] == hadm_id]
        else:
            resp_culture_stay = resp_culture.iloc[0:0]
            blood_culture_stay = blood_culture.iloc[0:0]

        micro_positive = (not resp_culture_stay.empty) or (not blood_culture_stay.empty)

        # ventilation >= 2 days before first criterion
        positive_event_times = []

        for df_ in [temp_stay, cough_stay, sputum_stay, rr_stay, spo2_stay, oxygen_stay, lung_stay]:
            if not df_.empty and "charttime" in df_.columns:
                positive_event_times.extend(df_["charttime"].dropna().tolist())

        first_criterion_time = min(positive_event_times) if len(positive_event_times) > 0 else pd.NaT

        vent_stay = vent[vent["stay_id"] == stay_id].copy()
        ventilation_ok = False

        if pd.notna(first_criterion_time) and not vent_stay.empty:
            for _, r in vent_stay.iterrows():
                st = r["starttime"]
                if pd.isna(st):
                    continue
                # đặt thở máy trước ít nhất 2 ngày
                if st <= first_criterion_time - pd.Timedelta(days=2):
                    ventilation_ok = True
                    break

        # rule gần đúng:
        # (>=2 triệu chứng) + imaging + ventilation
        # hoặc (>=3 triệu chứng) + ventilation
        vap_flag = (
            ((symptom_count >= 2) and imaging_positive and ventilation_ok) or
            ((symptom_count >= 3) and ventilation_ok) or
            ((symptom_count >= 2) and micro_positive and ventilation_ok)
        )

        results.append({
            "subject_id": subject_id,
            "hadm_id": hadm_id,
            "stay_id": stay_id,
            "symptom_count": symptom_count,
            "fever": symptom_hits["fever"],
            "cough": symptom_hits["cough"],
            "purulent_sputum": symptom_hits["purulent_sputum"],
            "tachypnea": symptom_hits["tachypnea"],
            "low_spo2": symptom_hits["low_spo2"],
            "oxygen_support": symptom_hits["oxygen_support"],
            "abnormal_lung_sounds": symptom_hits["abnormal_lung_sounds"],
            "imaging_positive": imaging_positive,
            "resp_culture_positive": not resp_culture_stay.empty,
            "blood_culture_positive": not blood_culture_stay.empty,
            "ventilation_ok_ge_2d_before_first_criterion": ventilation_ok,
            "first_criterion_time": first_criterion_time,
            "vap_suspected": vap_flag,
        })

    result_df = pd.DataFrame(results).sort_values(["stay_id"])

    if verbose:
        print("=== VAP check ===")
        print(result_df)

    return result_df
```
